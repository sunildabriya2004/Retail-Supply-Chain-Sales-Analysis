```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
# read dataset
retail_store = pd.read_csv('final_retail_store_clean_dataset.csv')
```

```python
total_sales = retail_store['Sales'].sum()
total_profit = retail_store['Profit'].sum()
profit_margin = total_profit / total_sales

print(f"total_sales is = {total_sales}\ntotal_profit is = {total_profit}\nprofit_margin is = {profit_margin}")
```

```python
# CATEGORY PERFORMANCE
retail_store.groupby('Category')[['Sales' , 'Profit']].sum().sort_values(by='Profit',ascending=False)
```

```python
# YEARLY SALES TREND
retail_store.groupby('order_year')[['Sales']].sum()
```

```python
# MONTHLY SALES TREND
retail_store.groupby(['order_month'])[['Sales']].sum()
```

```python
retail_store.columns
```

```python
# TOP 10 PRODUCTS BY SALES
retail_store.groupby('Product Name')[['Sales']].sum().sort_values(by='Sales' , ascending=False).head(10)
```

```python
# TOP 10 LOSS MAKING PRODUCTS BY SALES
retail_store.groupby(['Product Name'])[['Sales']].sum().sort_values(by='Sales').head(10)
```

```python
# HIGH SALES BUT LOSS
temp = retail_store.groupby('Sub-Category')[['Sales' , 'Profit']].sum()
temp[(temp['Sales'] > 100000) & (temp['Profit']<0)]
```

```python
# DISCOUNT IMPACT
retail_store[['Discount' , 'Profit']].corr()
```

```python
# SHIP MODE ANALYSIS
retail_store.groupby('Ship Mode')['Profit'].mean().sort_values(ascending=False)
```

```python
retail_store.groupby('Region')[['Sales' , 'Profit']].sum().sort_values(by='Sales')
```
